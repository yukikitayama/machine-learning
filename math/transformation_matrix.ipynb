# Transformation Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

## 2D rotation and stretching

```python
A = np.array([[1, -1], [2, 1]])
v = np.array([3, -2])
w = A @ v

plt.plot([0, v[0]], [0, v[1]], label='v')
plt.plot([0, w[0]], [0, w[1]], label='Av')
plt.grid()
plt.axis((-6, 6, -6, 6))
plt.legend()
plt.title('Rotation and stretching')
plt.show()
```

## 2D pure rotation

```python
theta = np.pi / 50
A = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
]) 
v = np.array([3, -2])
w = A @ v

plt.plot([0, v[0]], [0, v[1]], label='v')
plt.plot([0, w[0]], [0, w[1]], label='v')
plt.grid()
plt.axis((-4, 4, -4, 4))
plt.legend()
plt.title('Pure rotation')
plt.show()
```

## Pure rotation vs impure rotation

```python
v = np.array([3, -2])

thetas = np.linspace(0, 2 * np.pi, 100)
vector_magnitudes = np.zeros((len(thetas), 2))

for i in range(len(thetas)):
    
    theta = thetas[i]
    
    impure_rotation_matrix = np.array([
        [2 * np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    
    pure_rotation_matrix = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]        
    ])
    
    vector_magnitudes[i, 0] = np.linalg.norm(impure_rotation_matrix @ v)
    vector_magnitudes[i, 1] = np.linalg.norm(pure_rotation_matrix @ v)

plt.plot(thetas, vector_magnitudes)
plt.xlabel('Rotation angle (radian)')
plt.ylabel('Vector magnitude after rotation transformation')
plt.legend(['Impure rotation', 'Pure rotation'])
plt.show()
```

## Transform circle by various transformation matrice including singular matrix

### Identity matrix

```python
x = np.linspace(-np.pi, np.pi, 100)
xy = np.vstack((np.cos(x), np.sin(x)))
print(f'xy.shape: {xy.shape}')

transformation_matrix = np.array([
    [1, 0],
    [0, 1]
])

new_xy = transformation_matrix @ xy
print(f'new_xy.shape: {new_xy.shape}')

plt.plot(xy[0, :], xy[1, :], label='Original circle', marker='x')
plt.plot(new_xy[0, :], new_xy[1, :], label='Circle after transformation')
plt.axis('square')
plt.grid()
plt.show()
```

### Stretching matrix

```python
x = np.linspace(-np.pi, np.pi, 100)
xy = np.vstack((np.cos(x), np.sin(x)))
print(f'xy.shape: {xy.shape}')

transformation_matrix = np.array([
    [1, 0],
    [0, 2]
])

new_xy = transformation_matrix @ xy
print(f'new_xy.shape: {new_xy.shape}')

plt.plot(xy[0, :], xy[1, :], label='Original circle', marker='x')
plt.plot(new_xy[0, :], new_xy[1, :], label='Circle after transformation')
plt.axis('square')
plt.grid()
plt.show()
```

### Stretching and rotating matrix

```python
x = np.linspace(-np.pi, np.pi, 100)
xy = np.vstack((np.cos(x), np.sin(x)))
print(f'xy.shape: {xy.shape}')

transformation_matrix = np.array([
    [1, 0],
    [1, 2]
])

new_xy = transformation_matrix @ xy
print(f'new_xy.shape: {new_xy.shape}')

plt.plot(xy[0, :], xy[1, :], label='Original circle', marker='x')
plt.plot(new_xy[0, :], new_xy[1, :], label='Circle after transformation')
plt.axis('square')
plt.grid()
plt.show()
```

### Singular matrix
- Singular matrix is a matrix which has linear dependence

```python
x = np.linspace(-np.pi, np.pi, 100)
xy = np.vstack((np.cos(x), np.sin(x)))
print(f'xy.shape: {xy.shape}')

transformation_matrix = np.array([
    [1, 0],
    [1 * 2, 0 * 2]
])

new_xy = transformation_matrix @ xy
print(f'new_xy.shape: {new_xy.shape}')

plt.plot(xy[0, :], xy[1, :], label='Original circle', marker='x')
plt.plot(new_xy[0, :], new_xy[1, :], label='Circle after transformation')
plt.axis('square')
plt.grid()
plt.show()
```
