# Fourier Transform via Matrix Multiplication

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 52
omega = np.e**(-2 * np.pi * 1j / n)

F = np.zeros((n, n), dtype=complex)
for j in range(n):
    for k in range(n):
        m = j * k
        F[j, k] = omega ** m
        
plt.imshow(F.real, cmap='jet')
plt.show()

x = np.random.randn(n)
print('x')
print(x.shape)
print(x)
print()

X1 = F @ x
X2 = np.fft.fft(x)

plt.figure()
plt.plot(abs(X1))
plt.plot(abs(X2), marker='o', color='r', linestyle='None')
plt.show()
```
