# Logarithm

## Dependency

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
```

## Logarithmic function

```python
xs = np.linspace(0, 3, 100)
ys = np.log(xs)

plt.plot(xs, ys)
plt.title('Logarithm')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.grid()
plt.show()
```

## Math around log

In statistics, log is often used to describe the differences in probability between very small numbers. The above plot shows that for example, in 0.01 and 0.5, log(x) has a large different value, but for 2.5 and 3.0, log(x) doesn't change so much.

$\log(x)$ and $e^x$ are perfect inverse. Formula is the following

$$
\log(e^x) = x
$$
$$
e^{\log(x)} = x
$$
$$
\log(e^x) = e^{\log(x)}
$$

When taking exponent of the both sides of $\log(e^x) = x$,

$$
e^{\log(e^x)} = e^x
$$

So when we take exponent of log, it has an effect of taking out the input of the log.

```python
xs = np.linspace(0.0001, 1, 100)
log_e_x = np.log(np.exp(xs))

plt.plot(xs, log_e_x, label='log(e^x)', color='tab:blue')

xs = np.linspace(0.0001, 1, 10)
e_log_x = np.exp(np.log(xs))

plt.scatter(xs, e_log_x, label='e^{log(x)}', color='tab:orange')

plt.axis('square')
plt.legend()
plt.grid()
plt.show()
```

## Logarithmic scaling

- **Linear scaling** is by addition. For example, y axis spans by 0, 10, 20, 30, ... These are addition by 10.
- **Logarithmic scaling** is by multiplication. For example, y axis spans by 1, 10, 100, 100, ... Each tick is found by 10 multiplication to the previous tick $10^0, 10^1, 10^2, 10^3, ...$

The difference is visualized below. Linear scaling has a constant addition tick, and logarithmic scaling has 10 multiplication tick. In logarithmic scaling, $e^(x)$ becomes a straight line, because taking log of exponent has an offset effect as $\log(e^x) = x$. Use linear scaling unless you have reason to use log-plots. Use logarithmic scaling when the data has really big numbers happened such as in  physics, biology, or finance field.

```python
xs = np.linspace(1, 10, 100)
ys_lin = xs.copy()
ys_exp = np.exp(xs)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(xs, ys_lin, label='y = x')
plt.plot(xs, ys_exp, label='y = exp(x)')
plt.ylim(0, 100)
plt.title('Linear scaling')
plt.xlabel('x')
plt.ylabel('Linear scaling y')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.yscale('log')
plt.plot(xs, ys_lin, label='y = x')
plt.plot(xs, ys_exp, label='y = exp(x)')
plt.title('Logarithmic scaling')
plt.xlabel('x')
plt.ylabel('Logarithmic scaling y')
plt.legend()
plt.grid()

plt.show()
```
