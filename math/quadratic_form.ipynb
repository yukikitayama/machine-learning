## Big picture

**Quadratic form** is used for a **square** any size $m \times m$ matrix $S$. The matrix could be either **symmetric** or **non-symmetric** matrix. The square matrix is multipled with a column vector $w$. $w$ needs to have the same number of rows as the square matrix rows. As the following quadratic form shows, it always produces a **single number**.

$$
w^T S w
$$
$$
(1 \times m) (m \times m) (m \times 1) = (1 \times 1)
$$

It's taking a square matrix and pre and post multiply by some vector.

Quadratic form is used to find an interesting property of a given square matrix, and leads to **definiteness**

As mentioned, quadratic form works for more than $2 \times 2$ matrix, but computation will have too many characters and hard to interpret, so see the example of $2 \times 2$,

$$
S =
\begin{bmatrix}
a & b\\
c & d\\
\end{bmatrix}
\quad w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
ax + by\\
cx + dy
\end{bmatrix}
$$
$$
w^T S w = x(ax + by) + y(cx + dy)
$$
$$
w^T S w = ax^2 + bxy + cxy + dy^2
$$
$$
w^T S w = ax^2 + (b + c)xy + dy^2
$$

When $S$ is **symmetric**, transpose of a quadratic form gives the same result because,

$$
(w^T S w)^T
$$

By **LIVE EVIL rule**

$$
(w^T S w)^T = w^T S^T w^{TT}
$$

Because transpose of a transposed matrix is just an original matrix,

$$
(w^T S w)^T = w^T S^T w
$$

Because $S$ is a symmetric matrix and its transpose is the same,

$$
(w^T S w)^T = w^T S w
$$

$w^T S w$ is the original quadratic form.

What happens in math in $2 \times 2$ symmetric matrix is,

$$
S =
\begin{bmatrix}
a & b\\
b & d
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = ax^2 + 2bxy + dy^2
$$

When $S$ is an **identity matrix**, quadratic form gives a dot product of a vector with itself

$$
S =
\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = w^T I w = w^T w = x^2 + y^2
$$

```python
# Make a square matrix
S = np.array([
    [1, 3, -2],
    [0, 3, 4],
    [-5, -2, 4]
])

# Make a column vector
w = np.array([
    [-2],
    [4],
    [3]
])

# Quadratic form
qf = w.T @ S @ w

# Visualize
plt.subplot(231)
plt.imshow(w.T)
plt.axis('off')
plt.title('$w^T$')

plt.subplot(232)
plt.imshow(S)
plt.axis('off')
plt.title('S')

plt.subplot(233)
plt.imshow(w)
plt.axis('off')
plt.title('w')

plt.subplot(235)
plt.imshow(qf)
plt.axis('off')
plt.title('Quadratic form $w^T S w$')

plt.show()

print('w^T')
print(w.T)
print()
print('S')
print(S)
print()
print('w')
print(w)
print()
print('w^T S w')
print(qf)
print()
```

## Visualize quadratic form

```python
# Make several S matrices
k = 4
# S = np.zeros((4, ), dtype=np.object)
S = [None] * k
S[0] = np.array([
    [4, 4],
    [4, 9]
])
S[1] = np.array([
    [-4, -1],
    [3, -5]
])
S[2] = np.array([
    [0, 1],
    [2, 0]
])
S[3] = np.array([
    [1, 1],
    [1, 1]
])

# Make w vector
n = 30
w_range = np.linspace(-2, 2, n)

# Compute quadratic form
plt.figure(figsize=(10, 10))
qf = np.zeros((n, n))
for i in range(k):
    
    # Each w inpute
    for w1 in range(n):
        for w2 in range(n):
            
            # Make a current column vector w
            w = np.array([
                [w_range[w1]],
                [w_range[w2]]
            ])
            
            # Compute QR
            qf[w1, w2] = w.T @ S[i] @ w
            
    plt.subplot(2, 2, i + 1)
    plt.imshow(
        qf, 
        extent=[w_range[0], w_range[-1], w_range[0], w_range[-1]]
    )
    plt.colorbar()
    plt.xlabel('w1')
    plt.ylabel('w2')
    
    if i == 0:
        plt.title('All positive')
    elif i == 1:
        plt.title('All negative')
    elif i == 2:
        plt.title('Positive and negative')
    elif i == 3:
        plt.title('Positive and 0')

plt.show()
```

## Normalized quadratic form and eigenvector

Output of normalized quadratic form goes to infinity by $w$

**Normalized quadratic form** is modification to make the quadratic form useful

Used in PCA, and linear classification in statistics.

$$
\frac {w^T S w} {w^T w}
$$
$$
\frac {a x^2 + 2 b x y + d y^2} {x^2 + y^2}
$$

**Eigenvectors** point at peak and valley of the normalized quadratic form surface. For example, $2 \times 2$ matrix has 2 eigenvectors, and the associated normalized quadratic form also has 2 important directions. 

```python
# 3D visualization
mycmap = plt.get_cmap('gist_earth')

A = np.array([
    [-2, 3],
    [2, 8]
])

n = 30
xi = np.linspace(-2, 2, n)

X, Y = np.meshgrid(xi, xi)

qf = np.zeros((n, n))
qfN = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        
        x = np.array([
            [xi[i]],
            [xi[j]]
        ])
        
        qf[i, j] = x.T @ A @ x
        qfN[i, j] = (x.T @ A @ x) / (x.T @ x)

fig = plt.figure(figsize=(12, 7))

ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, qf.T, cmap=mycmap)
ax1.view_init(azim=-30, elev=30)
ax1.set_title('Non-normalized quadratic form')

ax1 = fig.add_subplot(122, projection='3d')
surf1 = ax1.plot_surface(X, Y, qfN.T, cmap=mycmap, antialiased=False)
ax1.view_init(azim=-30, elev=30)
ax1.set_title('Normalized quadratic form')

plt.show()
```

```python
A = np.array([
    [1, 2],
    [2, 3]
])

n = 30
xi = np.linspace(-2, 2, n)

X, Y = np.meshgrid(xi, xi)

qf = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        
        x = np.array([
            [xi[i]],
            [xi[j]]
        ])
        
        qf[i, j] = (x.T @ A @ x) / (x.T @ x)

# Eigendecomposition
_, V = np.linalg.eig(A)

# Scale up eigenvector for visualization
V *= 2

# Plot surface
plt.imshow(qf, extent=[-2, 2, -2, 2])

# Plot eigenvector
plt.plot([0, V[0, 0]], [0, V[1, 0]])
plt.plot([0, V[0, 1]], [0, V[1, 1]])

plt.show()
```

## Principal component analysis

If the matrix is the covariance matrix, eigendecomposition of the covariance matrix is principal component analysis (PCA).

Covariance matrix is $A A^T$ where $A$ is $m \times n$, and $m$ is the number of features, and $n$ is the number of data. So $A A^T$ is $m \times m$. And $A A^T$ is a symmetric matrix. 

If the given matrix is $n \times m$ where $n$ is the number of data and $m$ is the number of features, the covariance matrix is $m \times m$ $A^T A$.

The goal of PCA is to definte a linear combination of all the features such that the linear combination maximizes the covariance of the dataset.

In algebra, the goal of PCA is to define a weight $w_m$ for all the features, and use $S$ as the covariance matrix,

$$
w_m = \underset{w} {argmax} \{ \frac {w^T S w} {w^T w} \}
$$

```python
np.random.seed(0)

# Simulation parameters
# N: time points, number of data
# M: channels, number of features
N = 1000
M = 20

# Time vector (radian units)
t = np.linspace(0, 6 * np.pi, N)

print('t')
print(t.shape)
print()

plt.plot(t)
plt.title('t')
plt.show()

# Relationship across channels (imposing cvariance)
chanrel = np.sin(np.linspace(0, 2 * np.pi, M))

print('chanrel')
print(chanrel.shape)
print()

plt.plot(chanrel)
plt.title('chanrel')
plt.show()

# Initialize the given data
data = np.zeros((M, N))

for i in range(M):
    data[i, :] = np.sin(t) * chanrel[i]
    
print('data')
print(data.shape)
print()

plt.figure(figsize=(12, 5))
for i in range(M):
    plt.plot(t, data[i, :] + i * 2)
plt.title('Data before adding noise')
plt.xlabel('Time (a.u.)')
plt.ylabel('Channels')
plt.show()

# Add noise to data
data = data + np.random.randn(M, N) / 3

plt.figure(figsize=(12, 5))
for i in range(M):
    plt.plot(t, data[i, :] + i * 2)
plt.title('Data after adding noise')
plt.xlabel('Time (a.u.)')
plt.ylabel('Channels')
plt.show()


# Mean-center
for i in range(M):
    data[i, :] = data[i, :] - np.mean(data[i, :])
    
# Compute covariance matrix
covmat = (data @ data.T) / (N - 1)

print('Covariance matrix')
print(covmat.shape)
plt.imshow(covmat)
plt.xlabel('Channel')
plt.ylabel('Channel')
plt.title('Covariance matrix')
plt.show()
```

```python
np.random.seed(0)

# Principal component analysis
evals, evecs = np.linalg.eig(covmat)

# Sort eigenvalues and eigenvectors
idx = np.argsort(evals)[::-1]
evals = np.real(evals[idx])
evecs = evecs[:, idx]

# Convert eigenvalues into percent variance explained
evals = 100 * evals / np.sum(evals)

# Compute component time series
r = 2
comp_time_series = evecs[:, :r].T @ data

# Visualize
fig = plt.subplots(121, figsize=(10, 4))

# Eigenvalues
plt.subplot(121)
plt.plot(evals, 's-')
plt.xlabel('Component number')
plt.ylabel('$\lambda$ (% total variance)')
plt.title('Eigenspectrum')

# Principal components
plt.subplot(122)
plt.plot(evecs[:, 0], label='PC1')
plt.plot(evecs[:, 1], label='PC2')
plt.xlabel('Channel')
plt.ylabel('PC weight')
plt.title('Eigenvectors')
plt.legend()

plt.show()

# Original channel modulator
plt.plot(chanrel)
plt.xlabel('Channel')
plt.ylabel('Channel weight')
plt.title('Groud truth channel weights')

plt.show()

# Component time series
plt.plot(comp_time_series[0, :], label='PC1')
plt.plot(comp_time_series[1, :], label='PC2')
plt.xlabel('Time (a.u.)')
plt.ylabel('Activity')
plt.legend()
plt.title('Time course pltos')

plt.show()
```

## Definiteness

$A^T A$ is **square symmetric**, **invertible if $A$ is full column/row rank**, has **orthogonal eigenvectors**, and **real-valued eigenvalues**. 

Additionally, $A^T A$ is always either **positive definite** or **positive semi-definite**. It will never be **negative (semi)definite** and **indefinite**.

By substituting $A^T A$ to $S$ in **quadratic form**,

$$
w^T S w = w^T A^T A w
$$

By grouping them,

$$
w^T s w = (w^T A^T) (A w)
$$

By reverse **LIVE EVIL rule**,

$$
w^T S w = (A w)^T (A w)
$$

$A w$ is a column vector, so the right-hand side is the $v^T v$ dot product. Because $v^T v = ||v||^2$,

$$
w^T S w = ||A w||^2
$$

This is just a magnitude of a vector $A w$. It cannot be negative. So the quadratic form surface of $A^T A$ can only be **non-negative**.

**Singular matrix** has at least one 0 valued eigenvalue. $A^T A$ is a singular matrix if it doesn't have a full column rank. So when $A^T A$ is singular, $A^T A$ is **positive semidefinite**, but otherwise it's **positive definite**.

- $A^T A$ is **positive definite** when $A^T A$ is **invertible** (full column rank).
- $A^T A$ is **positive semidefinite** when $A^T A$ is **non-invertible** (reduced-rank).

It also works for $A A^T$. By substituting $A A^T$ to $S$,

$$
w^T S w = w^T A A^T w
$$

By grouping,

$$
w^T S w = (w^T A) (A^T w)
$$

By reverse **LIVE EVIL rule**,

$$
w^T S w = (w^T A) (w^T A)^T
$$

Because $w$ is a column vector, $w^T A$ is a row vector, $(w^T A)^T$ is a column vector, and row vector times column vector $(1 \times m) (m \times 1)$ is a scalar. So $(w^T A) (w^T A)^T$ is still a dot product,

$$
w^T S w = ||w^T A||^2
$$

And apply the same positive (semi)definite classification as $A^T A$ for this magnitude.

Not all symmetric matrices are positive (semi)definite. It could be negative. Above was saying that $A^T A$ is always positive (semi)definite where $A^T A$ happens to be symmetric.

### Eigenvalue

When a given matrix $A$ is positive-definite matrix, all the eigenvalues are positive.

Starting from eigendecomposition equation,

$$
A v = \lambda v
$$

Left-multiply by $v^T$ to make it a quadratic form

$$
v^T A v = v^T \lambda v
$$

Because $\lambda$ is a scalar, move it forward to change the order,

$$
v^T A v = \lambda v^T v
$$

Because $v^T v$ is a dot product,

$$
v^T A v = \lambda ||v||^2
$$

Because $A$ is positive definite, $v^T A v$ is always positive and $||v||^2$ is always positive because it's a squared magnitude, eigenvalue $\lambda$ needs to be positive.

When a given matrix $A$ is positive-semi-definite matrix, all the eigenvalues are non-negative.

By using the same math as above

$$
A v = \lambda v
$$
$$
v^T A v = v^T \lambda v
$$
$$
v^T A v = \lambda v^T v
$$
$$
v^T A v = \lambda ||v||^2
$$

Because $A$ is positive-semi-definite, $v^T A v$ is always non-negative, and $||v||^2$ is always positive, so eigenvalue needs to be either positive or zero (non-negative) to make the equation hold.
