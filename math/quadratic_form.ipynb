# Quadratic Form

## Dependency

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Big picture

**Quadratic form** is used for a **square** any size $m \times m$ matrix $S$. The matrix could be either **symmetric** or **non-symmetric** matrix. The square matrix is multipled with a column vector $w$. $w$ needs to have the same number of rows as the square matrix rows. As the following quadratic form shows, it always produces a **single number**.

$$
w^T S w
$$
$$
(1 \times m) (m \times m) (m \times 1) = (1 \times 1)
$$

It's taking a square matrix and pre and post multiply by some vector.

Quadratic form is used to find an interesting property of a given square matrix, and leads to **definiteness**

As mentioned, quadratic form works for more than $2 \times 2$ matrix, but computation will have too many characters and hard to interpret, so see the example of $2 \times 2$,

$$
S =
\begin{bmatrix}
a & b\\
c & d\\
\end{bmatrix}
\quad w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
ax + by\\
cx + dy
\end{bmatrix}
$$
$$
w^T S w = x(ax + by) + y(cx + dy)
$$
$$
w^T S w = ax^2 + bxy + cxy + dy^2
$$
$$
w^T S w = ax^2 + (b + c)xy + dy^2
$$

When $S$ is **symmetric**, transpose of a quadratic form gives the same result because,

$$
(w^T S w)^T
$$

By **LIVE EVIL rule**

$$
(w^T S w)^T = w^T S^T w^{TT}
$$

Because transpose of a transposed matrix is just an original matrix,

$$
(w^T S w)^T = w^T S^T w
$$

Because $S$ is a symmetric matrix and its transpose is the same,

$$
(w^T S w)^T = w^T S w
$$

$w^T S w$ is the original quadratic form.

What happens in math in $2 \times 2$ symmetric matrix is,

$$
S =
\begin{bmatrix}
a & b\\
b & d
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = ax^2 + 2bxy + dy^2
$$

When $S$ is an **identity matrix**, quadratic form gives a dot product of a vector with itself

$$
S =
\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = w^T I w = w^T w = x^2 + y^2
$$

```python
# Make a square matrix
S = np.array([
    [1, 3, -2],
    [0, 3, 4],
    [-5, -2, 4]
])

# Make a column vector
w = np.array([
    [-2],
    [4],
    [3]
])

# Quadratic form
qf = w.T @ S @ w

# Visualize
plt.subplot(231)
plt.imshow(w.T)
plt.axis('off')
plt.title('$w^T$')

plt.subplot(232)
plt.imshow(S)
plt.axis('off')
plt.title('S')

plt.subplot(233)
plt.imshow(w)
plt.axis('off')
plt.title('w')

plt.subplot(235)
plt.imshow(qf)
plt.axis('off')
plt.title('Quadratic form $w^T S w$')

plt.show()

print('w^T')
print(w.T)
print()
print('S')
print(S)
print()
print('w')
print(w)
print()
print('w^T S w')
print(qf)
print()
```

## Visualize quadratic form

```python
# Make several S matrices
k = 4
# S = np.zeros((4, ), dtype=np.object)
S = [None] * k
S[0] = np.array([
    [4, 4],
    [4, 9]
])
S[1] = np.array([
    [-4, -1],
    [3, -5]
])
S[2] = np.array([
    [0, 1],
    [2, 0]
])
S[3] = np.array([
    [1, 1],
    [1, 1]
])

# Make w vector
n = 30
w_range = np.linspace(-2, 2, n)

# Compute quadratic form
plt.figure(figsize=(10, 10))
qf = np.zeros((n, n))
for i in range(k):
    
    # Each w inpute
    for w1 in range(n):
        for w2 in range(n):
            
            # Make a current column vector w
            w = np.array([
                [w_range[w1]],
                [w_range[w2]]
            ])
            
            # Compute QR
            qf[w1, w2] = w.T @ S[i] @ w
            
    plt.subplot(2, 2, i + 1)
    plt.imshow(
        qf, 
        extent=[w_range[0], w_range[-1], w_range[0], w_range[-1]]
    )
    plt.colorbar()
    plt.xlabel('w1')
    plt.ylabel('w2')
    
    if i == 0:
        plt.title('All positive')
    elif i == 1:
        plt.title('All negative')
    elif i == 2:
        plt.title('Positive and negative')
    elif i == 3:
        plt.title('Positive and 0')

plt.show()
```

```python
# 3D visualization
mycmap = plt.get_cmap('gist_earth')

# Make several S matrices
k = 4
# S = np.zeros((4, ), dtype=np.object)
S = [None] * k
S[0] = np.array([
    [4, 4],
    [4, 9]
])
S[1] = np.array([
    [-4, -1],
    [3, -5]
])
S[2] = np.array([
    [0, 1],
    [2, 0]
])
S[3] = np.array([
    [1, 1],
    [1, 1]
])

# Make w vector
n = 30
w_range = np.linspace(-2, 2, n)

# Compute quadratic form
plt.figure(figsize=(10, 10))
qf = np.zeros((n, n))
for i in range(k):
    
    # Each w inpute
    for w1 in range(n):
        for w2 in range(n):
            
            # Make a current column vector w
            w = np.array([
                [w_range[w1]],
                [w_range[w2]]
            ])
            
            # Compute QR
            qf[w1, w2] = w.T @ S[i] @ w
            
    X, Y = np.meshgrid(w_range, w_range)
    Z = qf.T
    
    fig = plt.figure(figsize=(10, 6))
    ax1 = fig.add_subplot(221 + i, projection='3d')
    surf1 = ax1.plot_surface(X, Y, Z, cmap=mycmap)
    ax1.view_init(azim=-30, elev=30)

plt.show()
```

## Normalized quadratic form and eigenvector

Output of normalized quadratic form goes to infinity by $w$

**Normalized quadratic form** is modification to make the quadratic form useful

Used in PCA, and linear classification in statistics.

$$
\frac {w^T S w} {w^T w}
$$
$$
\frac {a x^2 + 2 b x y + d y^2} {x^2 + y^2}
$$

**Eigenvectors** point at peak and valley of the normalized quadratic form surface. For example, $2 \times 2$ matrix has 2 eigenvectors, and the associated normalized quadratic form also has 2 important directions. 

```python
# 3D visualization
mycmap = plt.get_cmap('gist_earth')

A = np.array([
    [-2, 3],
    [2, 8]
])

n = 30
xi = np.linspace(-2, 2, n)

X, Y = np.meshgrid(xi, xi)

qf = np.zeros((n, n))
qfN = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        
        x = np.array([
            [xi[i]],
            [xi[j]]
        ])
        
        qf[i, j] = x.T @ A @ x
        qfN[i, j] = (x.T @ A @ x) / (x.T @ x)

fig = plt.figure(figsize=(12, 7))

ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(X, Y, qf.T, cmap=mycmap)
ax1.view_init(azim=-30, elev=30)
ax1.set_title('Non-normalized quadratic form')

ax1 = fig.add_subplot(122, projection='3d')
surf1 = ax1.plot_surface(X, Y, qfN.T, cmap=mycmap, antialiased=False)
ax1.view_init(azim=-30, elev=30)
ax1.set_title('Normalized quadratic form')

plt.show()
```

```python
A = np.array([
    [1, 2],
    [2, 3]
])

n = 30
xi = np.linspace(-2, 2, n)

X, Y = np.meshgrid(xi, xi)

qf = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        
        x = np.array([
            [xi[i]],
            [xi[j]]
        ])
        
        qf[i, j] = (x.T @ A @ x) / (x.T @ x)

# Eigendecomposition
_, V = np.linalg.eig(A)

# Scale up eigenvector for visualization
V *= 2

# Plot surface
plt.imshow(qf, extent=[-2, 2, -2, 2])

# Plot eigenvector
plt.plot([0, V[0, 0]], [0, V[1, 0]])
plt.plot([0, V[0, 1]], [0, V[1, 1]])

plt.show()
```
