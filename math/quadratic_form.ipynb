# Quadratic Form

## Dependency

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Big picture

**Quadratic form** is used for a **square** any size $m \times m$ matrix $S$. The matrix could be either **symmetric** or **non-symmetric** matrix. The square matrix is multipled with a column vector $w$. $w$ needs to have the same number of rows as the square matrix rows. As the following quadratic form shows, it always produces a **single number**.

$$
w^T S w
$$
$$
(1 \times m) (m \times m) (m \times 1) = (1 \times 1)
$$

It's taking a square matrix and pre and post multiply by some vector.

Quadratic form is used to find an interesting property of a given square matrix, and leads to **definiteness**

As mentioned, quadratic form works for more than $2 \times 2$ matrix, but computation will have too many characters and hard to interpret, so see the example of $2 \times 2$,

$$
S =
\begin{bmatrix}
a & b\\
c & d\\
\end{bmatrix}
\quad w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w =
\begin{bmatrix}
x & y
\end{bmatrix}
\begin{bmatrix}
ax + by\\
cx + dy
\end{bmatrix}
$$
$$
w^T S w = x(ax + by) + y(cx + dy)
$$
$$
w^T S w = ax^2 + bxy + cxy + dy^2
$$
$$
w^T S w = ax^2 + (b + c)xy + dy^2
$$

When $S$ is **symmetric**, transpose of a quadratic form gives the same result because,

$$
(w^T S w)^T
$$

By **LIVE EVIL rule**

$$
(w^T S w)^T = w^T S^T w^{TT}
$$

Because transpose of a transposed matrix is just an original matrix,

$$
(w^T S w)^T = w^T S^T w
$$

Because $S$ is a symmetric matrix and its transpose is the same,

$$
(w^T S w)^T = w^T S w
$$

$w^T S w$ is the original quadratic form.

What happens in math in $2 \times 2$ symmetric matrix is,

$$
S =
\begin{bmatrix}
a & b\\
b & d
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = ax^2 + 2bxy + dy^2
$$

When $S$ is an **identity matrix**, quadratic form gives a dot product of a vector with itself

$$
S =
\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}
\quad
w =
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$
$$
w^T S w = w^T I w = w^T w = x^2 + y^2
$$

```python
# Make a square matrix
S = np.array([
    [1, 3, -2],
    [0, 3, 4],
    [-5, -2, 4]
])

# Make a column vector
w = np.array([
    [-2],
    [4],
    [3]
])

# Quadratic form
qf = w.T @ S @ w

# Visualize
plt.subplot(231)
plt.imshow(w.T)
plt.axis('off')
plt.title('$w^T$')

plt.subplot(232)
plt.imshow(S)
plt.axis('off')
plt.title('S')

plt.subplot(233)
plt.imshow(w)
plt.axis('off')
plt.title('w')

plt.subplot(235)
plt.imshow(qf)
plt.axis('off')
plt.title('Quadratic form $w^T S w$')

plt.show()

print('w^T')
print(w.T)
print()
print('S')
print(S)
print()
print('w')
print(w)
print()
print('w^T S w')
print(qf)
print()
```
