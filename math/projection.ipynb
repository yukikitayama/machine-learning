# Projection

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Project in R^2

```python
b = np.array([4, 1])
a = np.array([2, 5])

# Projection of b onto a and find beta
beta = (a.T @ b) / (a.T @ a)

print(f'Beta: {beta:.2f}')

# b
plt.plot(b[0], b[1], 'ko', label='b')
# a
plt.plot([0, a[0]], [0, a[1]], 'b', label='a')
# Projection of b onto a
plt.plot([b[0], beta * a[0]], [b[1], beta * a[1]], 'r--', label=r'b - $\beta$a')
# Beta scaled a
plt.plot(beta * a[0], beta * a[1], 'ko', label=r'$\beta a$')

plt.axis('square')
plt.grid()
plt.legend(loc='lower left')
plt.axis((-6, 6, -6, 6))
plt.show()
```

## Projection in R^N
- $Ax = b$, we know $A$ and $b$, and we wanna know $x$
- 2 ways to find it
  - Use math $x = (A^{T} A)^{-1}A^{T}b$
  - Use numpy.linalg.solve()

```python
m = 16
n = 10

# b
b = np.random.randn(m, 1)
print('b')
print(b.shape)
print(np.round(b, 2))
print()

# A
A = np.random.randn(m, n)
print('A')
print(A.shape)
print(np.round(A, 2))
print()

# Math
x1 = np.linalg.inv(A.T @ A) @ (A.T @ b)
print('x by math')
print(x1.shape)
print(np.round(x1, 2))
print()

# Numpy
x2 = np.linalg.solve(A.T @ A, A.T @ b)
print('x by numpy')
print(x2.shape)
print(np.round(x2, 2))
```

```python
m = 3
n = 2

# Vector b
b = np.random.randn(m, 1)

# maxtrix A
A = np.random.randn(m, n)

# Solution
x = np.linalg.solve(A.T @ A, A.T @ b)
Ax = A @ x

print('b')
print(np.round(b, 6))
print()

print('Ax')
print(np.round(b, 6))
print()

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.gca(projection='3d')

# Plot vector b
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'r', label='b')

# Plot vector Ax
ax.plot([0, Ax[0]], [0, Ax[1]], [0, Ax[2]], 'b', label='Ax')

# Plot projection line
ax.plot(
    [Ax[0][0], b[0]],
    [Ax[1][0], b[1]],
    [Ax[2][0], b[2]],
    'g',
    label='Projection'
)

# Draw a plane of column space of A
xx, yy = np.meshgrid(np.linspace(-2, 2), np.linspace(-2, 2))
cp = np.cross(A[:, 0], A[:, 1])
z1 = (-cp[0] * xx - cp[1] * yy) * 1. / cp[2]
ax.plot_surface(xx, yy, z1, alpha=.4)

plt.legend()
plt.show()
```

## Decompose a vector
- to a **parallel vector** to the reference vector
- to a **orthogonal vector** to a reference vector 

```python
# W vector to be decomposed
w = np.array([2, 3])
print('w')
print(w)
print()

# V vector as reference
v = np.array([4, 0])
print('v')
print(v)
print()

# W parallel to v, scaled version of v
beta = np.dot(v, w) / np.dot(v, v)
w_parallel_to_v = beta * v
print('w parallel to v')
print(w_parallel_to_v)
print()

# W orthogonal to v
w_orthogonal_to_v = w - w_parallel_to_v
print('w orthogonal to v')
print(w_orthogonal_to_v)
print()

# Confirm it was correct
# By sum of parallel and orthogonal becomes w
print('By sum')
print((w_parallel_to_v + w_orthogonal_to_v) - w)
print()

# By dot product of paralle and orthogonal is 0
print('By dot product')
print(np.dot(w_parallel_to_v, w_orthogonal_to_v))

# Plot w to be decomposed
plt.plot([0, w[0]], [0, w[1]], 'r', linewidth=3)

# Plot v to be referenced
plt.plot([0, v[0]], [0, v[1]], 'b', linewidth=1.5)

# Plot w-parallel-to-v vector
plt.plot([0, w_parallel_to_v[0]], [0, w_parallel_to_v[1]], 'r--', linewidth=3)

# Plot w-orthogonal-to-v vector
plt.plot([0, w_orthogonal_to_v[0]], [0, w_orthogonal_to_v[1]], 'r:', linewidth=3)

plt.legend(['w', 'v', '$w_{||v}$', '$w_{\perp v}$'])
plt.axis('square')
plt.grid()
plt.axis([-5, 5, -5, 5])
plt.show()
```
