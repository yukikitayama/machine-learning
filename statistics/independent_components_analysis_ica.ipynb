# Independent Components Analysis (ICA)

A dimension reduction procedure.

Find a **weighted combination of features** to find out **as least Gaussian distributed as possible** 

## Assumption

- It uses **central limit theorem**
- Signals are non-Gaussian distributed
- Random mixture of signals are Gaussian distributed
- ICA assumes Gaussian distributions are bad

## Algorithm

- Whiten the data to remove covariance to remove linear dependencies
- Rotate axes to minimize shared information

## Goal

- From the mixture of the data, recover the original data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
```

```python
n = 1000

# Non-gaussian
dist1 = np.random.rand(n)

# Non-gaussian
dist2 = np.random.rand(n)**4

plt.hist(dist1, 100)
plt.show()

plt.hist(dist2, 100)
plt.show()

plt.hist(dist1 + dist2, 100)
plt.show()
```

```python
data = np.vstack(
    (0.4 * dist1 + 0.3 * dist2, 0.8 * dist1 - 0.7 * dist2)
)

fastica = FastICA(max_iter=10000, tol=0.0000001)
b = fastica.fit_transform(data)
iscores = b @ data

plt.hist(iscores[0, :], 100)
plt.show()

plt.hist(iscores[1, :], 100)
plt.show()
```

```python
plt.plot(dist1, iscores[0, :], 'o')
plt.show()

plt.plot(dist2, iscores[1, :], 'o')
plt.show()
```
