# QQ Plot

## Dependency

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

QQ plot stands for quintile-quintile plot.

```python
n = 1000
data = np.random.randn(n)
x = np.linspace(-4, 4, n)

# Create probability density function
theonorm = stats.norm.pdf(x)
theonorm = theonorm / max(theonorm)

# Create histogram from data and normalize
yy, xx = np.histogram(data, 40)
yy = yy / max(yy)
xx = (xx[:-1] + xx[1:]) / 2

plt.plot(xx, yy, label='Empirical')
plt.plot(x, theonorm, label='Theoretical')
plt.legend()
plt.show()
```

```python
# zscore() transforms the data such that mean of 0 and standard diviation of 1
# to make empirical data and analytic data on the same scale
z_sort_data = np.sort(stats.zscore(data))
sort_normal = stats.norm.ppf(np.linspace(0, 1, n))

# Make QQ plot
plt.plot(sort_normal, z_sort_data, 'o')
x_l, x_r = plt.xlim()
y_l, y_r = plt.ylim()
lims = [
    np.min([x_l, x_r, y_l, y_r]),
    np.max([x_l, x_r, y_l, y_r])
]
plt.xlim(lims)
plt.ylim(lims)
# Draw comparison line
plt.plot(lims, lims)
plt.xlabel('Theoretical normal')
plt.ylabel('Observed data')
plt.title('QQ plot')
plt.axis('square')
plt.show()
```
