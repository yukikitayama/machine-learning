# Normal Distribution

## Dependency

```python
import numpy as np
import matplotlib.pyplot as plt
import math
plt.style.use('default')
```

The formula of normal distribution is,

$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2} (\frac{x - \mu}{\sigma})^2}
$$

```python
mu = 0
sigma = 1


def normal(x, mu, sigma):
    return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp((-1 / 2) * ((x - mu) / sigma)**2)


xs = np.linspace(-3, 3, 100)
ys = [normal(x, mu, sigma) for x in xs]


plt.plot(xs, ys)
plt.title('Normal distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Boiling down the formula of normal distribution to the essential part gives us,

$$
y = e^{-x^2}
$$

```python
def simple_normal(x):
    return math.exp(-x**2)


xs = np.linspace(-3, 3, 100)
ys = [simple_normal(x) for x in xs]


plt.plot(xs, ys)
plt.title('Simple normal distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
