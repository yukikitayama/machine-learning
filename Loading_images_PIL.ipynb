```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import scipy
from PIL import Image
```

```python
img = mpimg.imread('coffee_picture.jpg')
```

```python
plt.imshow(img)
```

```python
image_arr = np.array(img)
image_arr = np.rot90(image_arr, k = 3)
plt.imshow(image_arr)
plt.show()
```

```python
img = Image.open('coffee_picture.jpg')
size = 100
img = img.resize((size, size))
plt.imshow(img)
plt.show()
```

```python
from PIL import Image
img = Image.open('coffee_picture.jpg')
size = 100
img = img.resize((size, size))
img = img.rotate(270)
plt.imshow(img)
plt.show()
```
