# Bayesian Statistics

## Raining in Seattle

You are about to get on a plane to Seattle. You want to know if you should bring an umbrella. You call 3 random friends of yours who live there and ask each independently if it’s raining. Each of your friends has a $\frac{2}{3}$ chance of telling you the truth and a $\frac{1}{3}$ chance of messing with you by lying. All 3 friends tell you that “Yes” it is raining. What is the probability that it’s actually raining in Seattle?

All 3 friends said yes it's raining, but each has $\frac{1}{3}$ probability to lie, and lie means not raining. So we have the following probability of not raining in Seattle.

$$
P(\text{Not raining in Seattle}) = \frac{1}{3} \times \frac{1}{3} \times \frac{1}{3}
$$
$$
= \frac{1}{27}
$$

We have either raining or not raining in Seattle. We wanna know the probability of raining in Seattle, but we can rephrase the probability of raining in Seattle as 1 minus the probability of not raining in Seattle. 

$$
P(\text{Raining in Seattle}) = 1 - P(\text{Not raining in Seattle})
$$
$$
= 1 - \frac{1}{27}
$$
$$
= \frac{26}{27}
$$

We solve the same problem by Bayesian approach. Let $R$ denote the probability of raining in Seattle, and $F_i$ denote the probability that $i$th friend tells us it's raining. Because all 3 friends told us yes and we wanna know the probability of raining in Seattle, the probability we want is,

$$
P(R | F_1 \cap F_2 \cap F_3)
$$

By **Beyes' theorem** we have,

$$
P(A | B) = \frac{P(B | A)P(A)}{P(B)}
$$

So in our problem, it's written as,

$$
P(R | F_1 \cap F_2 \cap F_3) = \frac{P(F_1 \cap F_2 \cap F_3 | R) P(R)}{P(F_1 \cap F_2 \cap F_3)}
$$

We use $P(R) = \frac{1}{2}$, because we have either raining or not raining in Seattle.

$P(F_1 \cap F_2 \cap F_3 | R)$ is the probability that, given it's raining, all 3 friends says it's raining, that is each friend is telling the truth and each probability is $\frac{2}{3}$ as stated in the question, so the probability is,

$$
P(F_1 \cap F_2 \cap F_3 | R) = (\frac{2}{3})^3
$$
$$
= \frac{8}{27}
$$

Law of total probability says, assuming $B$ is binary and $B^c$ is inverse of $B$,

$$
P(A) = P(A | B) P(B) + P(A | B^c) P(B^c)
$$

Applyinh the above law of total probability to $P(F_1 \cap F_2 \cap F_3)$ is, 

$$
P(F_1 \cap F_2 \cap F_3) = P(F_1 \cap F_2 \cap F_3 | R) P(R) + P(F_1 \cap F_2 \cap F_3 | R^c) P(R^c)
$$

Because we assumed $P(R) = \frac{1}{2}$ and raining or not raining is binary,

$$
P(R^c) = \frac{1}{2}
$$

$P(F_1 \cap F_2 \cap F_3 | R^c)$ is the probability that, given it's not raining, all 3 friends say yes it's raining, that is all 3 friends are lying, and each lying probability is $\frac{1}{3}$ as stated in the question,

$$
P(F_1 \cap F_2 \cap F_3 | R^c) = (\frac{1}{3})^3
$$
$$
= \frac{1}{27}
$$

Putting them together,

$$
P(F_1 \cap F_2 \cap F_3) = P(F_1 \cap F_2 \cap F_3 | R) P(R) + P(F_1 \cap F_2 \cap F_3 | R^c) P(R^c)
$$
$$
= \frac{8}{27} \times \frac{1}{2} + \frac{1}{27} \times \frac{1}{2}
$$
$$
= \frac{9}{54}
$$
$$
= \frac{1}{6}
$$

Putting everything together,

$$
P(R | F_1 \cap F_2 \cap F_3) = \frac{P(F_1 \cap F_2 \cap F_3 | R) P(R)}{P(F_1 \cap F_2 \cap F_3)}
$$
$$
= \frac{\frac{8}{27} \frac{1}{2}}{\frac{1}{6}}
$$
$$
= \frac{\frac{8}{27}}{\frac{1}{3}}
$$
$$
= \frac{8}{9}
$$
